```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
path = "../Risultati/outlier_detection.csv"
df = pd.read_csv(path)
df.head()
df.reset_index(drop=True)
```

```python
df_z = df[(df['method'] == 'z')]
df_z
```

```python
df_l = df[(df['method'] == 'lof')]
df_l
```

```python
df_i = df[(df['method'] == 'iforest')]
df_i
```

```python
df_h = df[(df['method'] == 'hst')]
df_h
```

```python
plt.plot(df_z['percentage'], df_z['f1'], label='Z-Score')
plt.plot(df_l['percentage'], df_l['f1'], label='LOF')
plt.plot(df_i['percentage'], df_i['f1'], label='IForest')
plt.plot(df_h['percentage'], df_h['f1'], label='HST')
plt.legend()
```

```python
plt.plot(df_z['percentage'], df_z['precision'], label='Z-Score')
plt.plot(df_l['percentage'], df_l['precision'], label='LOF')
plt.plot(df_i['percentage'], df_i['precision'], label='IForest')
plt.plot(df_h['percentage'], df_h['precision'], label='HST')
plt.legend()
```

```python
plt.plot(df_z['percentage'], df_z['recall'], label='Z-Score')
plt.plot(df_l['percentage'], df_l['recall'], label='LOF')
plt.plot(df_i['percentage'], df_i['recall'], label='IForest')
plt.plot(df_h['percentage'], df_h['recall'], label='HST')
plt.legend()
```

```python
path = "../Risultati/outlier_detection_regression.csv"
df = pd.read_csv(path)
df.head()
df.reset_index(drop=True)
df['percentage'] = 100 - df['percentage']
```

```python
df
```

```python
df_none = df[(df['method'] == 'none')]
df_none
```

```python
df_lof = df[(df['method'] == 'lof')]
df_lof
```

```python
df_z = df[(df['method'] == 'z')]
df_z
```

```python
df_iforest = df[(df['method'] == 'iforest')]
df_iforest
```

```python
df_hst = df[(df['method'] == 'hst')]
df_hst
```

```python
perc = [90,80,70,60,50]
plt.plot(perc, [df_none['r2'].iloc[0]]*len(perc), 'r--', label='Clean')
plt.plot(df_none['percentage'].iloc[1:],df_none['r2'].iloc[1:], '--', label='No action')
plt.plot(df_z['percentage'],df_z['r2'], label='z')
plt.plot(df_lof['percentage'],df_lof['r2'], label='LOF')
plt.plot(df_iforest['percentage'],df_iforest['r2'], label='IForest')
plt.plot(df_hst['percentage'],df_hst['r2'], label='HST')
plt.title("Regression Performance")
plt.legend()
plt.plot()
```
