```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_tabular = pd.read_csv("../../../Risultati/Electrical_TABULAR/outlier_detection.csv")
df_tabular['percentage'] = 100 - df_tabular['percentage']
df_stream_c = pd.read_csv("../../../Risultati/NEWeather/outlier_detection.csv")
df_stream_c['percentage'] = 100 - df_stream_c['percentage']
df_stream_r = pd.read_csv("../../../Risultati/Air_Quality/outlier_detection.csv")
df_stream_r['percentage'] = 100 - df_stream_r['percentage']
```

```python
df_stream_c
```

```python
df_t_z = df_tabular[(df_tabular['method'] == 'z')].groupby(['percentage']).mean(numeric_only=True).reset_index()
df_t_z
```

```python
df_c_z = df_stream_c[(df_stream_c['method'] == 'z')].groupby(['percentage']).mean(numeric_only=True).reset_index()
df_c_z
```

```python
df_r_z = df_stream_r[(df_stream_r['method'] == 'z')].groupby(['percentage']).mean(numeric_only=True).reset_index()
df_r_z
```

```python
plt.plot(df_t_z['percentage'],df_t_z['f1'],label='Tabular')
plt.plot(df_c_z['percentage'],df_c_z['f1'],label='NEWeather')
plt.plot(df_r_z['percentage'],df_r_z['f1'],label='AirQuality')
plt.title('F1-Score in outlier recognition - Z-score')
plt.legend()
```
