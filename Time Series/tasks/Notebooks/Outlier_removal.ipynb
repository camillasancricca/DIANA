```python
import numpy as np
import pandas as pd
import sklearn 
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, recall_score, f1_score
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn import metrics

```

```python
df_clean = pd.read_csv("../../Datasets/ChlorineConcentration.csv", sep="  ", header= None)
df_clean
```

```python
from scipy.stats import zscore
cols = df_clean.columns[1:]
z=df_clean[cols[1:]].apply(zscore)
z
```

```python
z=[z>3]
z
```

```python
z = np.array(z)
z = np.squeeze(z)

z.shape
```

```python
outlier_mask = np.any(z,axis=1)
print(outlier_mask)
```

```python
df_filtered_Z = df_clean[~outlier_mask]
df_filtered_Z.describe()
```

```python
df_filtered_Z = df_filtered_Z.reset_index(drop=True)
df_filtered_Z
```

```python
from pyod.models.iforest import IForest

ifor = IForest()
#df_o.drop(['date_time'],axis=1,inplace=True)
ifor = ifor.fit(df_clean[cols[1:]])
outliers = ifor.predict(df_clean[cols[1:]]).astype(np.bool_)

#outliers = np.where(outliers == 1)[0]
print(outliers)
```

```python
df_filtered_IF = df_clean[~outliers]
df_filtered_IF.describe()
```

```python
df_filtered_IF = df_filtered_IF.reset_index(drop=True)
```
